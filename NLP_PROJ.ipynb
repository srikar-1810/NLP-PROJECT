```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')
```

```python
df = pd.read_csv("cyberbullying_tweets.csv")
df.head()
```

```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words("english"))

def clean_text(text):
    if not isinstance(text, str):
        return ""

    text = text.lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"@\w+", "", text)
    text = re.sub(r"[^a-z\s]", " ", text)
    words = text.split()

    cleaned = [
        lemmatizer.lemmatize(w)
        for w in words
        if w not in stop_words
    ]

    return " ".join(cleaned)
```

```python
df = df.rename(columns={
    'tweet_text': 'text',
    'cyberbullying_type': 'label'
})

df['clean_text'] = df['text'].apply(clean_text)

df.head()
```

```python
cleaned_path = "/content/drive/MyDrive/cleaned_cyberbullying.csv"
df.to_csv(cleaned_path, index=False)

cleaned_path
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report

X = df['clean_text']
y = df['label']

tfidf = TfidfVectorizer(max_features=6000)
X_vec = tfidf.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_vec, y, test_size=0.2, random_state=42)

model = LinearSVC()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
import pickle

model_path = "/content/drive/MyDrive/svm_model.pkl"
tfidf_path = "/content/drive/MyDrive/tfidf.pkl"

pickle.dump(model, open(model_path, "wb"))
pickle.dump(tfidf, open(tfidf_path, "wb"))

model_path, tfidf_path
```

```python
import pickle

model = pickle.load(open("/content/drive/MyDrive/svm_model.pkl", "rb"))
tfidf = pickle.load(open("/content/drive/MyDrive/tfidf.pkl", "rb"))

def predict(text):
    cleaned = clean_text(text)
    vec = tfidf.transform([cleaned])
    return model.predict(vec)[0]

print(predict("I hate you so much"))
print(predict("Have a great day"))
```
